```python
import numpy as np
import os
import tabulate
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline

import load_data
import models
import curves
import utils
```

```python
def plot_curves(paths, train=True):
    fig = plt.figure(figsize=(8,4))
    for path in paths:
        arrays = np.load(path)
        value, label = ('tr_nll', 'Train loss') if train else ('te_err', 'Test error')
        #value, label = ('tr_err', 'Train error') if train else ('te_err', 'Test error')
        #print(arrays[value])
        xnew = np.linspace(arrays['ts'].min(), arrays['ts'].max(), 300) #300 represents number of points to make between T.min and T.max
        spl = make_interp_spline(arrays['ts'], arrays[value], k=3) #BSpline object
        power_smooth = spl(xnew)
        plt.plot(xnew,power_smooth)
        
        #plt.yticks([0, 0.1, 0.2, 0.4, 0.8, 1.6])
        plt.xlabel('t')
        plt.ylabel(label)
        plt.legend(['Segment', 'Plolychain', 'Bezier'])
    plt.savefig('train_loss.jpg' if train else 'test_loss.jpg')
    plt.show()
```

```python
#paths = ['./chain/chain1.npz', './eval_polychain/curve3.npz', './eval_bezier/curve1.npz']
paths=['./chain/chain1_2.npz', './eval_polychain/curve1_2.npz', './eval_bezier/curve1_2.npz']
```

```python
plot_curves(paths, train=True)
```

```python
plot_curves(paths, train=False)
```
